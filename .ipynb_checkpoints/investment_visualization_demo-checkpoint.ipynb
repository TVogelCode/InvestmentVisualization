# Personal Investment Portfolio Analysis
This Python script is designed for the analysis of personal investments. It imports trading data from CSV files, which contain securities transactions downloaded from the Fidelity trading platform. In addition, it extracts price data for all the securities listed in the CSV files using the AlphaVantage API.

With this data, the script constructs a Pandas dataframe. The dataframe contains the value and profitability of each security in the portfolio over time. Moreover, it calculates the total value and profit of the entire investment portfolio, as well as the allocation of the investments by region over time.

The region data for all securities is imported as an excel-file that was created manually. The collection of the region data might be automated in the future but is currently not supported by the AlphaVantage API.

## Environment Setup

```python
# Import necessary libraries for data manipulation and analysis
import pandas as pd
import numpy as np

# Import libraries for data visualization
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns

# Import utility libraries
import os
import requests
import time
import locale

# Imports for Demo
import csv
import random
import datetime

# Set the matplotlib style to 'ggplot' for better visualization
plt.style.use('ggplot')

# Configure pandas options to display all columns and rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Get the AlphaVantage API key from environment variables
api_key = os.environ.get('API_KEY')

# Check if the API key is available; if not, raise an error
if api_key is None:
    raise ValueError('API_KEY environment variable not found. Please set it before running the notebook.')
```

## Data Import and Cleaning
### Historical Price Data
#### Pulling Data from the AlphaVantage API

```python
# Get unique values from the 'Symbol' column in 'trades_df' to identify all securities in the portfolio
# Define your list of securities
securities = ['IDEV', 'IEUR', 'IEMG', 'ITOT']

# Convert the unique symbols to a set for efficient manipulation
symbols_set = set(securities)
```

```python
# Function to retrieve historical price data for a set of stock symbols from AlphaVantage
def get_stock_prices(symbols, api_key):
    """
    Retrieve historical price data for a set of stock symbols from AlphaVantage and store it in a dictionary.

    Args:
        symbols (set): Set of stock symbols.
        api_key (str): AlphaVantage API key.

    Returns:
        dict: A dictionary with stock symbols as keys and DataFrames with price data as values.
    """
    stock_prices = {}

    base_url = 'https://www.alphavantage.co/query'
    function = 'TIME_SERIES_DAILY'
    output_size = 'full'

    for symbol in symbols:
        # Build the URL with symbol and API key
        params = {
            'function': function,
            'symbol': symbol,
            'outputsize': output_size,
            'apikey': api_key
        }
        url = f'{base_url}?function={function}&symbol={symbol}&outputsize={output_size}&apikey={api_key}'

        try:
            r = requests.get(url)
            r.raise_for_status()  # Raise an exception for HTTP errors
            data = r.json()

            if 'Time Series (Daily)' in data:
                # Extract and process the time series data
                time_series_data = data['Time Series (Daily)']
                df = pd.DataFrame.from_dict(time_series_data, orient='index')
                df = df.reset_index()
                df = df.rename(columns={'index': 'Date', '1. open': 'Open', '2. high': 'High', '3. low': 'Low',
                                       '4. close': 'Close', '5. volume': 'Volume'})

                float_columns = ['Open', 'High', 'Low', 'Close', 'Volume']
                df[float_columns] = df[float_columns].astype(float)
                df['Date'] = pd.to_datetime(df['Date'])

                # Sort the DataFrame by date
                df = df.sort_values(by='Date', ascending=True)

                stock_prices[symbol] = df
            else:
                print(f"No data found for symbol: {symbol}")

            # Add a delay to avoid API rate limiting
            time.sleep(15)
        except requests.exceptions.RequestException as e:
            print(f"Error retrieving data for symbol {symbol}: {e}")

    return stock_prices

# Call the function to retrieve stock prices for the symbols in 'symbols_set' using the 'api_key'
stock_prices = get_stock_prices(symbols_set, api_key)
```

#### Data Cleaning

```python
# Reduce the data in 'stock_prices' to the relevant timeframe
# Find the date of the first and last trade in the 'trades_df'
first_trade_date = datetime.date(2016, 8, 1)
last_trade_date = datetime.date(2023, 9, 30)

# Filter the 'stock_prices' dictionary for each stock symbol to contain only the relevant data
filtered_stock_prices = {
    key: df.query('@first_trade_date <= Date <= @last_trade_date').reset_index(drop=True)
    for key, df in stock_prices.items()
}
```

### Trading Data
#### Generate random Trade Data for Demo of Script

```python
def generate_random_trade(symbol, current_date):
    if symbol == 'CASH':
        quantity = 550
        price = 1
        amount = quantity
    else:
        quantity = random.randint(1, 5)
        symbol_df = filtered_stock_prices.get(symbol)
        if symbol_df is not None:
            date_match = symbol_df[symbol_df['Date'].astype(str) == current_date.strftime('%Y-%m-%d')]
            if not date_match.empty:
                price = date_match['Close'].values[0]
            else:
                price = None  # Handle the case where the date doesn't exist in the DataFrame
        else:
            price = None  # Handle the case where the symbol is not found in the DataFrame
        if price is None:
            amount = None
        else:
            amount = -quantity * price
    settlement_date = current_date + datetime.timedelta(days=2)
    
        
    return {
        'Run Date': current_date.strftime('%Y-%m-%d'),
        'Action': 'BUY',
        'Symbol': symbol,
        'Quantity': quantity,
        'Currency': 'USD',
        'Price': price,
        'Amount': amount,
        'Settlement Date': settlement_date.strftime('%Y-%m-%d')
    }

def write_trades_to_csv(trades):
    with open('trades.csv', mode='w', newline='') as file:
        fieldnames = ['Run Date', 'Action', 'Symbol', 'Quantity', 'Currency', 'Price', 'Amount', 'Settlement Date']
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(trades)

def main():
    securities = ['IDEV', 'IEUR', 'IEMG', 'ITOT']
    start_date = first_trade_date
    end_date = last_trade_date
    trades = []

    current_date = start_date
    while current_date <= end_date:
        if current_date.day in [1, 15]:
            for symbol in securities:
                trades.append(generate_random_trade(symbol, current_date))
            trades.append(generate_random_trade('CASH', current_date))
        current_date += datetime.timedelta(days=1)

    write_trades_to_csv(trades)
    print("Trades have been generated and saved to 'trades.csv'.")

if __name__ == "__main__":
    main()
```

#### Import Demo CSV File and generate DataFrame from it

```python
# Define the folder path where CSV files are located
folder_path = '/Users/tim/Documents/coding_projects/investments'

# Initialize an empty list to store individual DataFrames
dataframes = []

# Iterate through the files in the folder
for filename in os.listdir(folder_path):
    # Check if the file has a .csv extension
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        try:
            # Read each CSV file into a DataFrame and append it to the list
            df = pd.read_csv(file_path)
            dataframes.append(df)
        except Exception as e:
            print(f"Error reading {filename}: {str(e)}")

# Check if any CSV files were found and processed
if dataframes:
    # Concatenate all DataFrames in the list into one
    trades_df = pd.concat(dataframes, ignore_index=True)
    print("CSV files successfully concatenated.")
else:
    print("No CSV files found in the folder.")
```

#### DataFrame Cleaning

```python
# Define a dictionary to map old column names to new ones
column_name_mapping = {
    'Run Date': 'Date',
    'Security Description': 'Security_Description',
    'Security Type': 'Security_Type',
    'Exchange Quantity': 'Exchange_Quantity',
    'Exchange Currency': 'Exchange_Currency',
    'Exchange Rate': 'Exchange_Rate',
    'Accrued Interest': 'Accrued_Interest',
    'Settlement Date': 'Settlement_Date'
}

# Rename the columns in the trades_df DataFrame using the dictionary
trades_df = trades_df.rename(columns=column_name_mapping)
```

```python
# Remove leading and trailing whitespace from 'Action' and 'Symbol' columns
trades_df['Action'] = trades_df['Action'].str.strip()
trades_df['Symbol'] = trades_df['Symbol'].str.strip()
#trades_df['Price'] = trades_df['Price'].str.strip()
#trades_df['Amount'] = trades_df['Amount'].str.strip()

# Replace 'Symbol' value with 'CASH' where 'Action' is 'Electronic Funds Transfer Received (Cash)'
#trades_df.loc[trades_df['Action'] == 'Electronic Funds Transfer Received (Cash)', 'Symbol'] = 'CASH'
```

```python
# Define columns that need their data type changed
datetime_columns = ['Date', 'Settlement_Date']
category_columns = ['Symbol', 'Currency']
#float_columns = ['Price', 'Amount']

# Convert selected columns to the correct data types
trades_df[datetime_columns] = trades_df[datetime_columns].apply(pd.to_datetime)
trades_df[category_columns] = trades_df[category_columns].astype('category')
#trades_df['Price'] = trades_df['Price'].astype('float')
#trades_df['Amount'] = trades_df['Amount'].astype('float')
```

```python
# Sort the DataFrame by the 'Date' column from oldest to newest
trades_df = (
    trades_df
    .dropna(subset=['Price'])
    .sort_values(by='Date', ascending=True)
    .reset_index(drop=True)
)
```

### Region Data

```python
# Read data from an Excel file into the info_df DataFrame
info_df = pd.read_excel('security_info.xlsx', sheet_name='Info')

# Convert the 'Type' column to a category data type for efficient storage
info_df['Type'] = info_df['Type'].astype('category')
```

## Aggregating all Data into the final DataFrame

```python
# Create a date range based on the dates of the first and last trade 
date_range = pd.date_range(start=first_trade_date, end=last_trade_date, freq='D')

# Create an empty DataFrame with 'Date' column
portfolio_value = pd.DataFrame({'Date': date_range})

# Process data for each symbol in the portfolio
for symbol in symbols_set:
    stock_data = filtered_stock_prices[symbol]
    
    # Filter the DataFrame for the specific symbol
    filtered_df = trades_df[trades_df['Symbol'] == symbol]
    
    # Calculate the cumulative sum of Quantity and Amount for the symbol
    cumulative_sum_df = pd.DataFrame({
        'Date': filtered_df['Date'],
        f'Cum_Quantity_{symbol}': filtered_df['Quantity'].cumsum(),
        f'Cum_Amount_{symbol}': filtered_df['Amount'].cumsum()
    })
    
    # Group by 'Date' and select the last row for each date
    cumulative_sum_df = cumulative_sum_df.groupby('Date').tail(1).reset_index(drop=True)
    
    # Merge the stock's 'Close' price with the portfolio DataFrame
    portfolio_value = portfolio_value.merge(stock_data[['Date', 'Close']], on='Date', how='left')
    portfolio_value = portfolio_value.rename(columns={'Close': f'Close_{symbol}'})
    portfolio_value = portfolio_value.merge(cumulative_sum_df, on='Date', how='left')
    
    # Fill NaN values for days with no trades
    portfolio_value[f'Cum_Quantity_{symbol}'] = portfolio_value[f'Cum_Quantity_{symbol}'].fillna(method='ffill').fillna(0)
    portfolio_value[f'Cum_Amount_{symbol}'] = portfolio_value[f'Cum_Amount_{symbol}'].fillna(method='ffill').fillna(0)
    portfolio_value[f'Close_{symbol}'] = portfolio_value[f'Close_{symbol}'].fillna(method='ffill').fillna(0)

# Process data for 'CASH' deposits
filtered_df = trades_df[trades_df['Symbol'] == 'CASH']

# Calculate the cumulative sum of Amount for 'CASH'
cumulative_sum_df = pd.DataFrame({
    'Date': filtered_df['Date'],
    f'Cum_Amount_CASH': filtered_df['Amount'].cumsum()
})

# Group by 'Date' and select the last row for each date
cumulative_sum_df = cumulative_sum_df.groupby('Date').tail(1).reset_index(drop=True)

# Merge 'CASH' data with the portfolio DataFrame
portfolio_value = portfolio_value.merge(cumulative_sum_df, on='Date', how='left')

# Fill NaN values for days with no trades
portfolio_value[f'Cum_Amount_CASH'] = portfolio_value[f'Cum_Amount_CASH'].fillna(method='ffill').fillna(0)

# Calculate 'Total_Value' and 'Total_Amount' of the portfolio
portfolio_value['Total_Value'] = sum(
    portfolio_value[f'Close_{symbol}'] * portfolio_value[f'Cum_Quantity_{symbol}']
    for symbol in symbols_set
)
portfolio_value['Total_Amount'] = sum(
    portfolio_value[f'Cum_Amount_{symbol}']
    for symbol in symbols_set
)

# Calculate 'Total_Profit' of the portfolio
portfolio_value['Total_Profit'] = portfolio_value['Total_Value'] + portfolio_value['Total_Amount']

# Calculate value, profit, and percentage profit for each symbol
for symbol in symbols_set:
    portfolio_value[f'Value_{symbol}'] = portfolio_value[f'Close_{symbol}'] * portfolio_value[f'Cum_Quantity_{symbol}']
    portfolio_value[f'Profit_{symbol}'] = portfolio_value[f'Value_{symbol}'] + portfolio_value[f'Cum_Amount_{symbol}']
    portfolio_value[f'%_Profit_{symbol}'] = portfolio_value[f'Profit_{symbol}'] / portfolio_value['Total_Profit']

# Calculate regional exposure for each symbol and store it in respective columns
regions = ['NA', 'EU', 'AS', 'ME', 'LA', 'AF']

for region in regions:
    portfolio_value[f'Total_{region}'] = sum(
        portfolio_value[f'Cum_Quantity_{symbol}'] *
        portfolio_value[f'Close_{symbol}'] /
        portfolio_value['Total_Value'] *
        info_df.loc[info_df['Symbol'] == symbol, region].values[0]
        if len(info_df.loc[info_df['Symbol'] == symbol, 'NA'].values) > 0
        else np.nan
        for symbol in symbols_set
    )
```

### Currently Invested and Uninvested Money in Account

```python
# Set the locale to 'en_US.UTF-8' for proper currency formatting
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')

# Calculate the total money invested and the cash balance in the account
last_trade_date_data = portfolio_value[portfolio_value['Date'] == portfolio_value['Date'].max()]
money_invested = -last_trade_date_data['Total_Amount'].values[0]
cash_in_account = (last_trade_date_data['Cum_Amount_CASH'].values[0] - money_invested)

# Format the results with thousands separators and two decimal places
formatted_money_invested = locale.currency(money_invested, grouping=True)
formatted_cash_in_account = locale.currency(cash_in_account, grouping=True)

# Print the formatted results
print(f'Money invested: {formatted_money_invested}')
print(f'Cash in account: {formatted_cash_in_account}')
```

### Region Allocation on Date of last Trade

```python
# Define custom labels for the regions (in the same order as 'regions')
custom_labels = ['North America', 'Europe', 'Latin America', 'Africa', 'Asia']

# Set a threshold for label display (e.g., 5%)
label_threshold = 5

# Extract data for the last date in the portfolio
last_date_data = portfolio_value.iloc[-1]

# Select the columns representing region allocations for the pie chart
regions = ['Total_NA', 'Total_EU', 'Total_LA', 'Total_AF', 'Total_AS']
allocation = last_date_data[regions]

# Create a pie chart to visualize region allocation on the last date
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(allocation, labels=None, startangle=90,
                                   autopct=lambda p: f'{p:.0f}%' if p > label_threshold else '')
plt.title('Region Allocation on Date of last Trade')
plt.axis('equal')  # Ensure the chart is a perfect circle

# Add labels for significant portions
for label, autotext in zip(custom_labels, autotexts):
    if autotext.get_text() != '':  # Skip labels with no percentage
        autotext.set_text(f'{label} ({autotext.get_text()})')

plt.show()
```

### ROI in % over Time

```python
# Calculate the cumulative return on investment (ROI)
roi = (portfolio_value['Total_Value'] + portfolio_value['Total_Amount']) / -portfolio_value['Total_Amount'] * 100

# Calculate the total money invested and the final value of investments
last_trade_date_data = portfolio_value[portfolio_value['Date'] == portfolio_value['Date'].max()]
money_invested = -last_trade_date_data['Total_Amount'].values[0]
value_investments = last_trade_date_data['Total_Value'].values[0]

# Calculate the annualized ROI
roi_per_year = (value_investments - money_invested) / money_invested * 100 / (
    (last_trade_date - first_trade_date).days) * 365

# Create a figure and axis for the ROI plot
plt.figure(figsize=(10, 6))
ax = plt.gca()

# Plot the cumulative ROI as a line graph for uninvested money
plt.plot(portfolio_value['Date'], roi, marker='', linestyle='-', label='Uninvested Money')

# Set labels and a title for the plot
plt.xlabel('Date')
plt.ylabel('ROI [%]')
plt.title('Return on Investment (ROI) Over Time')

# Add grid lines to improve visualization
plt.grid(True)

# Display a legend for the plot
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Ensure a tight layout for improved spacing
plt.tight_layout()

# Show the ROI plot
plt.show()

# Print the annualized ROI
print(f'Annualized ROI: {roi_per_year:.2f}%')
```
